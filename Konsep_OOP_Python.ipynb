```
# latihan 3
class Hero:
    def __init__(self, name, hp, attack_power):
        self.name = name
        self.hp = hp
        self.attack_power = attack_power

    def info(self):
        print(f"Hero: {self.name} | HP: {self.hp} | Power: {self.attack_power}")

    def serang(self, lawan):
        print(f"{self.name} menyerang {lawan.name}!")
        lawan.diserang(self.attack_power)

    def diserang(self, damage):
        self.hp -= damage
        print(f"{self.name} terkena damage {damage}. Sisa HP: {self.hp}")


# ======================
# Child Class (Inheritance)
# ======================
class Mage(Hero):
    def __init__(self, name, hp, attack_power, mana):
        # Memanggil constructor Parent (Hero)
        super().__init__(name, hp, attack_power)
        self.mana = mana

    def info(self):
        print(f"{self.name} [Mage] | HP: {self.hp} | Mana: {self.mana}")

    # Skill khusus Mage
    def skill_fireball(self, lawan):
        if self.mana >= 20:
            print(f"{self.name} menggunakan Fireball ke {lawan.name}!")
            self.mana -= 20
            lawan.diserang(self.attack_power * 2)
        else:
            print(f"{self.name} gagal skill! Mana tidak cukup.")

# ======================
# Main Program
# ======================
print("\n--- Update Class Hero ---")

eudora = Mage("Eudora", 80, 200, 100)
balmond = Hero("Balmond", 200, 10)

eudora.info()
eudora.serang(balmond)        # Method warisan dari Hero
eudora.skill_fireball(balmond)

#Analisis
# error nya adlah AttributeError: 'Mage' object has no attribute 'name'
# Alasannya: Saat `super().__init__(name, hp, attack_power)` dihapus, constructor milik **Hero (class induk)** tidak dijalankan. Akibatnya atribut `name`, `hp`, dan `attack_power` **tidak pernah dibuat**, walaupun nilai `"Eudora"` dikirim saat instansiasi.
#Peran `super()`:`super()` berfungsi untuk **memanggil constructor class induk**, sehingga class anak (`Mage`) mewarisi dan menyimpan data (`name`, `hp`, `attack_power`) dari class induk (`Hero`). Tanpa `super()`, data dari induk tidak terhubung ke anak.
```

```
#latihan 4
# ==================================
# Class Hero (Encapsulation)
# ==================================
class Hero:
    def __init__(self, name, hp, attack_power):
        self.name = name
        self.__hp = hp              # PRIVATE attribute
        self.attack_power = attack_power

    # Getter
    def get_hp(self):
        return self.__hp

    # Setter (AWALNYA DENGAN VALIDASI)
    def set_hp(self, nilai_baru):
        if nilai_baru < 0:
            print("HP tidak boleh negatif!")
        else:
            self.__hp = nilai_baru

    def info(self):
        print(f"Hero: {self.name} | HP: {self.__hp} | Power: {self.attack_power}")


# ==================================
# Main Program
# ==================================
hero1 = Hero("Layla", 100, 15)

print("\n--- INFO AWAL ---")
hero1.info()

# ==================================
# 1. PERCOBAAN HACKING (NAME MANGLING)
# ==================================
print("\n--- PERCOBAAN HACKING ---")
print(f"Mencoba akses paksa: {hero1._Hero__hp}")

# ==================================
# 2. UJI VALIDASI SETTER
# (SIMULASI: LOGIKA VALIDASI DIHAPUS)
# ==================================

# Setter TANPA validasi (MENIRU KESALAHAN)
def set_hp_tanpa_validasi(self, nilai_baru):
    self.__hp = nilai_baru

# Mengganti method set_hp dengan versi tanpa validasi
Hero.set_hp = set_hp_tanpa_validasi

print("\n--- UJI SETTER TANPA VALIDASI ---")
hero1.set_hp(-100)   # HP negatif
hero1.info()

#Analisis
#1. Percobaan Hacking (Name Mangling) Nilai HP tetap muncul, tidak terjadi error. Hal ini karena Python menggunakan name mangling, yaitu mengubah __hp menjadi _Hero__hp agar atribut tidak mudah diakses dari luar class.
#Walaupun bisa diakses, cara ini tidak dianjurkan karena melanggar prinsip encapsulation dan berisiko merusak data.

#2. Uji Validasi Setter
#Saat hero1.set_hp(-100) dijalankan, HP hero menjadi -100.
#Ini menunjukkan bahwa tanpa validasi, data bisa menjadi tidak masuk akal.
#Method setter sangat penting untuk menjaga integritas data, agar nilai HP tetap valid dan sesuai aturan game.
```

```
#tugas analisis 4
from abc import ABC, abstractmethod

# 1. Abstract Class / Interface
# (Blueprint / Kontrak)

class GameUnit(ABC):

    @abstractmethod
    def serang(self, target):
        pass

    @abstractmethod
    def info(self):
        pass

# 2. Implementasi pada Class Hero

class Hero(GameUnit):
    def __init__(self, nama):
        self.nama = nama

    # Wajib implementasi method serang
    def serang(self, target):
        print(f"Hero {self.nama} menebas {target}!")

    # Wajib implementasi method info
    def info(self):
        print(f"Saya adalah Hero: {self.nama}")

# 3. Implementasi pada Class Monster

class Monster(GameUnit):
    def __init__(self, jenis):
        self.jenis = jenis

    # Implementasi serang versi Monster
    def serang(self, target):
        print(f"Monster {self.jenis} menggigit {target}!")

    # Implementasi info versi Monster
    def info(self):
        print(f"Saya adalah Monster: {self.jenis}")

h = Hero("Alucard")
m = Monster("Serigala")

h.info()
m.info()

h.serang("Monster")
m.serang("Hero")

#Analisis
#1. Melanggar Kontrak
#Muncul error:
#TypeError: Can't instantiate abstract class Hero with abstract method serang
#Artinya Hero tidak lengkap karena tidak mengimplementasikan method yang diwajibkan oleh GameUnit.

#2. Abstract Class
#GameUnit tidak bisa dibuat objek karena hanya berfungsi sebagai blueprint/kontrak untuk class turunan agar strukturnya sama.

```

```
#Tugas Proyek
from abc import ABC, abstractmethod

# ==================================
# ABSTRACT CLASS (Abstraction)
# ==================================
class BarangElektronik(ABC):
    def __init__(self, nama, harga_dasar):
        self.nama = nama
        self.__stok = 0                  # private attribute
        self.__harga_dasar = harga_dasar # private attribute

    # Getter
    def get_stok(self):
        return self.__stok

    # Method tambah stok (Setter + Validasi)
    def tambah_stok(self, jumlah):
        if jumlah < 0:
            print(f"Gagal update stok {self.nama}! Stok tidak boleh negatif ({jumlah}).")
        else:
            self.__stok += jumlah
            print(f"Berhasil menambahkan stok {self.nama}: {self.__stok} unit.")

    # Getter harga dasar (dipakai child)
    def _get_harga_dasar(self):
        return self.__harga_dasar

    # Abstract Methods
    @abstractmethod
    def tampilkan_detail(self):
        pass

    @abstractmethod
    def hitung_harga_total(self, jumlah):
        pass


# ==================================
# CHILD CLASS: Laptop (Inheritance)
# ==================================
class Laptop(BarangElektronik):
    def __init__(self, nama, harga_dasar, processor):
        super().__init__(nama, harga_dasar)
        self.processor = processor

    def tampilkan_detail(self):
        pajak = self._get_harga_dasar() * 0.10
        print(f"[LAPTOP] {self.nama} | Proc: {self.processor}")
        print(f" Harga Dasar: Rp {self._get_harga_dasar():,} | Pajak(10%): Rp {int(pajak):,}")

    def hitung_harga_total(self, jumlah):
        harga = self._get_harga_dasar()
        pajak = harga * 0.10
        return (harga + pajak) * jumlah


# ==================================
# CHILD CLASS: Smartphone (Inheritance)
# ==================================
class Smartphone(BarangElektronik):
    def __init__(self, nama, harga_dasar, kamera):
        super().__init__(nama, harga_dasar)
        self.kamera = kamera

    def tampilkan_detail(self):
        pajak = self._get_harga_dasar() * 0.05
        print(f"[SMARTPHONE] {self.nama} | Cam: {self.kamera}")
        print(f" Harga Dasar: Rp {self._get_harga_dasar():,} | Pajak(5%): Rp {int(pajak):,}")

    def hitung_harga_total(self, jumlah):
        harga = self._get_harga_dasar()
        pajak = harga * 0.05
        return (harga + pajak) * jumlah


# ==================================
# POLYMORPHISM FUNCTION
# ==================================
def proses_transaksi(daftar_barang):
    print("\n--- STRUK TRANSAKSI ---")
    total = 0

    for i, (barang, jumlah) in enumerate(daftar_barang, start=1):
        print(f"{i}. ", end="")
        barang.tampilkan_detail()
        subtotal = barang.hitung_harga_total(jumlah)
        print(f" Beli: {jumlah} unit | Subtotal: Rp {int(subtotal):,}\n")
        total += subtotal

    print("----------------------------------------")
    print(f"TOTAL TAGIHAN: Rp {int(total):,}")
    print("----------------------------------------")


# ==================================
# MAIN PROGRAM (User Story)
# ==================================
print("--- SETUP DATA ---")

laptop1 = Laptop("ROG Zephyrus", 20_000_000, "Ryzen 9")
hp1 = Smartphone("iPhone 13", 15_000_000, "12MP")

laptop1.tambah_stok(10)
hp1.tambah_stok(-5)
hp1.tambah_stok(20)

# User membeli barang
keranjang = [
    (laptop1, 2),
    (hp1, 1)
]

proses_transaksi(keranjang)
```
